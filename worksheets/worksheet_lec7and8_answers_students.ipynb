***Question 1.1:*** Connect via pymongo.

```python
from pymongo import MongoClient
import json
import urllib.parse

with open('data/credentials_mongodb.json') as f:
    login = json.load(f)

username = login['username']
password = urllib.parse.quote(login['password'])
host = login['host']
url = "mongodb+srv://{}:{}@{}/?retryWrites=true&w=majority".format(username, password, host)
```

```python
client = MongoClient(url)
```

***Question 3:*** 


Write a query to select fields `title`,`num_mflix_comments` and `year` for all documents with `num_mflix_comments` to be 2. Also sort the result by year in ascending order.

Hint: When asked for certain fields, it is like 'select col1, col2'

```python
list(
    client['sample_mflix']['movies'].find( 
        filter={'num_mflix_comments': 2},
        projection={'title':1,'num_mflix_comments':1,'year':1,'_id':0},
        sort=[('year',1)]
)
)
```

***Question 1:*** 
(4 point)

Write a query to count the number of movies casted either  'Tommy Lee Jones' or 'Will Smith' with `num_mflix_comments` greater than 120.

Hint: Make sure you select 'type' as 'movie' - this is applicable for all questions where we are asking about movies.

```python
client['sample_mflix']['movies'].count_documents(
    filter={
'cast': {
    '$in': ['Tommy Lee Jones',
    'Will Smith']
}    ,
'type':{'$eq': 'movie'},
'num_mflix_comments': {'$gt': 120}}
)
```

```python
client['sample_mflix']['movies'].count_documents(
    filter={
'cast': {
    '$in': ['Tommy Lee Jones',
    'Will Smith']
},
'type':'movie',
'num_mflix_comments': {'$gt': 120}}
)
```

***Question 2:*** 

Write a query to count the number of movies casted either 'Tommy Lee Jones' or 'Will Smith' with `num_mflix_comments` greater than 120 or `runtime` less than 100 min.

Hint: You can develop on the Question 1 query to include the `or` condition.

```python
client['sample_mflix']['movies'].count_documents(
    filter={
'cast': {
    '$in': ['Tommy Lee Jones',
    'Will Smith']
},
'type':{'$eq': 'movie'},
'$or' : [{'num_mflix_comments': {'$gt': 120}},
    {'runtime': {'$lt': 100}}
]
    }
)
```

***Question 3:*** 

Write a query to select distinct movie titles casted either 'Tommy Lee Jones' or (corrected from and --> or) 'Will Smith' with `nominations` greater than or equal to 10.

Hint: You can develop on the Question 1 query.

```python
list(client['sample_mflix']['movies'].find(
    filter={
'cast': {
    '$in': ['Tommy Lee Jones',
    'Will Smith']
},
'type':{'$eq': 'movie'},
'awards.nominations':{'$gte' : 10}
    }
).distinct('title')
)
```

***Question 4:*** 

Write a query to count the number of movies casted 'Norman Kerry' with `tomatoes.viewer.meter` greater than or equal to 5 and the field `languages` exists. 

Note: Make sure that you don't include documents which have a blank space in their `tomatoes.viewer.meter` field! ('$ne': '')

```python
client['sample_mflix']['movies'].count_documents(
    filter={
'cast': {
    '$in': ['Norman Kerry']
}    ,
'languages':{'$exists': True},
'tomatoes.viewer.meter': {'$gte' : 5,
'$ne': ''}
    }
)
```
