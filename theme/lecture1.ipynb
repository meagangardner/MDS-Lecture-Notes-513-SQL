## Let's talk about DISTINCT keyword ?

````python
%%sql

SELECT
     continent
FROM
    country
;
````

````python
%%sql

SELECT
    DISTINCT continent
FROM
    country
;
````

````python
%%sql

SELECT
    DISTINCT continent, region
FROM
    country
;
````

## How can we sort the data that is returned ?

````python
%%sql

SELECT
    name, population
FROM
    country
ORDER BY
    population
;
````

````python
%%sql

SELECT
    name, population
FROM
    country
ORDER BY
    population DESC
;
````

## I don't want to return all of the data! What can I do?

````python
%%sql

SELECT
    name, continent
FROM
    country
LIMIT
    5
;
````

The order of SQL clauses does matter. The correct order for the ones that we’ve learned so far is:

        SELECT
        FROM
        WHERE
        ORDER BY
        LIMIT

### How to uninstall Postgres on Windows:

Uninstall the program, stop the services, delete all residual files + remove all registry keys under the name “postgres” or “pgadmin”, and you should be able to setup a new password.

---

### How to uninstall Postgres on Mac:

- Open a terminal window, and run these commands

cd /Library/PostgreSQL/15
open uninstall-postgresql.app/

This should launch the uninstall program. Note that the target folder in step 1 (e.g. 15) depends on the version you have.

- After uninstalling, go to this path: /Library/ and delete the folder PostgreSQL. This folder may be locked; if that’s the case, right click on it => “Get info” => scroll down and click the orange/brown lock icon. Also, click the three dots ... and click “Apply to enclosed items”.

- Open a terminal and run the following commands:

sudo dscl . -create /Users/postgres UserShell /bin/sh
sudo dscl . -create /Users/postgres NFSHomeDirectory /Library/PostgreSQL

- Restart your computer, and reinstall Postgres.

---

- By default, the server group appearing in the left pane of pgAdmin is named PostgreSQL <X>. Regardless, it doesn’t matter at all what you call your server group on pgAdmin; that’s up to you.

---

### There are in general two sets of passwords we have to deal with:

- First, One master password for unlocking pgAdmin interface. This password is set the first time you open pgAdmin. This doesn’t have anything to do with your database user accounts.

- Second, There’s a username and password which is associated with your database account. The postgres username on your local Postgres installation, which is created by default, is an example of that. This is for accessing the databases whenever you make a connection to the database, e.g. from Jupyter, psql, etc.. You set a password for this user (i.e. postgres user) during Postgres installation.
